```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
books = pd.read_csv("Amazon book sales/Books.csv")
rating = pd.read_csv("Amazon book sales/Ratings.csv")
users = pd.read_csv("Amazon book sales/Users.csv")
```

```python
books.describe(include="all")
books["ISBN length"] = books["ISBN"].astype(str).str.len()
books.head(3)
```

```python
filtered_isbn = books[(books["ISBN length"]!=10) & (books["ISBN length"]!=13)]
filtered_isbn
```

```python
books["Year-Of-Publication"] = pd.to_numeric(books["Year-Of-Publication"], errors='coerce')
books.loc[(books["Year-Of-Publication"] < 2007) & (books["ISBN length"]>10), 'ISBN'] = books.loc[(books["Year-Of-Publication"] < 2007) & (books["ISBN length"]>10), 'ISBN'].str[:10]
books["ISBN length"] = books['ISBN'].str.len()
filtered_isbn = books[(books["ISBN length"]!=10) & (books["ISBN length"]!=13)]
filtered_isbn
```

```python
books_without_img = books[["ISBN","Book-Title", "Book-Author", "Year-Of-Publication", "Publisher"]]
books_without_img.head(3)
```

```python
rating.head(4)
```

```python

rating["ISBN"] = rating["ISBN"].str.replace(r'[^0-9X]','', regex=True)
rating["ISBN"] = rating["ISBN"].str.strip()
rating["ISBN length"] = rating["ISBN"].astype(str).str.len()
filtered_isbn_rate = rating[(rating["ISBN length"]!=10) & (rating["ISBN length"]!=13)]
```

```python
filtered_isbn_rate
```

```python
rating['ISBN_valid'] = rating['ISBN'].str.len().isin([10,13])
invalid_ISBN = rating[~rating['ISBN_valid']]
invalid_ISBN.head(2)
```

```python
rating = rating[rating['ISBN_valid']]
rating.info()
```

```python
books_rating = pd.merge(books, rating, on='ISBN', how = 'inner')
books_rating.head(1)
```

```python
books_rating_without_img = books_rating[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher', 'User-ID', 'Book-Rating']]
books_rating_without_img.head(2)
```

```python
users.head(4)
```

```python
full_table = pd.merge(books_rating_without_img, users, on='User-ID', how = 'inner')
full_table.head(2)
```

```python
summary_rating = full_table.groupby(['ISBN','Book-Title'])['Book-Rating'].sum().reset_index()
top_10_books = summary_rating.sort_values('Book-Rating', ascending=False).head(10)
top_10_books
```

```python

plt.figure(figsize=(12,6))
sns.barplot(
    x = "Book-Rating",
    y = "Book-Title",
    data=top_10_books,
    palette = "viridis")
plt.xlabel('Rating')
plt.ylabel('Books')
plt.title('Top 10 Amazon books')
plt.show()
```

```python
summary_rating = full_table.groupby('Book-Author')['Book-Rating'].sum().reset_index()
top_10_authors = summary_rating.sort_values('Book-Rating', ascending=False).head(10)
top_10_authors
```

```python
plt.figure(figsize=(12,6))
sns.barplot(
    x = "Book-Rating",
    y = "Book-Author",
    data=top_10_authors,
    palette = "pastel")
plt.xlabel('Rating')
plt.ylabel('Author')
plt.title('Top 10 Amazon sold Authors')
plt.show()
```

```python
user_rating_count = full_table.groupby('User-ID')['Book-Rating'].count().reset_index()
user_rating_count.rename(columns={'Book-Rating':'Rating_Count'}, inplace=True)
top_10_users = user_rating_count.sort_values('Rating_Count', ascending=False).head(10)
top_10_users["User-ID"] = top_10_users["User-ID"].astype(str)
top_10_users
```

```python
plt.figure(figsize=(12,6))
sns.barplot(
    x = "Rating_Count",
    y = "User-ID",
    data=top_10_users,
    palette = "viridis")
plt.xlabel('Rating coint')
plt.ylabel('User ID')
plt.title('Top 10 Active users')
plt.show()
```

```python
bins = [18,28,38,48,58,68,78,88,98]
labels = ['18-28', '29-38', '39-48', '49-58', '59-68', '69-78', '79-88', '89-98']
full_table['Age_group']=pd.cut(full_table['Age'], bins=bins, labels= labels, right=True)
group_by_age = full_table.groupby('Age_group')['Book-Rating'].count().reset_index()
group_by_age.rename(columns={'Book-Rating':'Rating_Count'}, inplace=True)
group_by_age
```

```python
full_table['Year-Of-Publication'] = full_table['Year-Of-Publication'].astype('Int64')
invalid_years = full_table[(full_table['Year-Of-Publication']< 1500) | (full_table['Year-Of-Publication']> 2025)]
invalid_years['Book-Title'].nunique()
valid_years= full_table[(full_table['Year-Of-Publication']>= 1500) & (full_table['Year-Of-Publication']< 2025)]
valid_years
```

```python
bins = list(range(1800, 2026, 10))
labels = [f"{b}-{b+9}" for b in bins[:-1]]

valid_years["Decade"] = pd.cut(valid_years['Year-Of-Publication'], bins=bins, labels=labels, right=False)

decade_rating = valid_years.groupby("Decade")["Book-Rating"].mean().reset_index()

plt.figure(figsize=(12,6))
sns.barplot(
    x="Decade",
    y="Book-Rating",
    data=decade_rating,
    palette="viridis"
)
plt.xlabel("Decade")
plt.ylabel("Average Rating")
plt.title("Average Book Rating by Decade")
plt.xticks(rotation=45)
plt.show()
```

```python
year_avg = valid_years.groupby("Year-Of-Publication")["Book-Rating"].mean().reset_index()

plt.figure(figsize=(15,6))
sns.lineplot(
    x="Year-Of-Publication",
    y="Book-Rating",
    data=year_avg,
    marker="o"
)
plt.xlabel("Year")
plt.ylabel("Average Rating")
plt.title("Average Book Rating by Year")
plt.show()
```

```python
summary_rating_by_publisher = full_table.groupby('Publisher')['Book-Rating'].sum().reset_index()
top_10_publishers = summary_rating_by_publisher.sort_values('Book-Rating', ascending=False).head(10)
top_10_publishers
plt.figure(figsize=(12,6))
sns.barplot(
    x = "Book-Rating",
    y = "Publisher",
    data=top_10_publishers,
    palette = "magma")
plt.xlabel('Book-Rating')
plt.ylabel('Publisher')
plt.title('Top 10 Publisher collected rating')
plt.show()
```

```python
most_popular_book_by_publisher = full_table.groupby(['Publisher','Book-Title'])['Book-Rating'].sum().reset_index()
top_10_most_popular_book_by_publisher = most_popular_book_by_publisher.sort_values('Book-Rating', ascending=False).head(10)
top_10_most_popular_book_by_publisher["Publisher_title"] = top_10_most_popular_book_by_publisher['Publisher'] + '-' + top_10_most_popular_book_by_publisher['Book-Title']

plt.figure(figsize=(12,6))
sns.barplot(
    x = "Book-Rating",
    y = "Publisher_title",
    data=top_10_most_popular_book_by_publisher,
    palette = "magma")
plt.xlabel('Book-Rating')
plt.ylabel('Publisher - Book Title')
plt.title('Top 10 Books by Total rating per Publisher')
plt.show()
```
